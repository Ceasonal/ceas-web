```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('cyberForm.csv')

# Data cleaning

# Convert Timestamp column to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Clean 'Level of familiarity with cybersecurity concepts' column
df['Level of familiarity with cybersecurity concepts'] = df['Level of familiarity with cybersecurity concepts'].str.strip().str.capitalize()

# Clean 'Cybersecurity topics interested in' column
topics_mapping = {
    'Password Security': 'Password Security',
    'Network Security': 'Network Security',
    'Malware Protection': 'Malware Protection',
    'Data Encryption': 'Data Encryption'
}
df['Cybersecurity topics interested in'] = df['Cybersecurity topics interested in'].apply(lambda x: ', '.join([topics_mapping.get(topic.strip(), '') for topic in x.split(',')]))

# Clean 'Type of devices used most frequently' column
df['Type of devices used most frequently'] = df['Type of devices used most frequently'].str.strip().str.capitalize()

# Convert 'Have you ever fallen victim to a phishing attack?' column to boolean
df['Have you ever fallen victim to a phishing attack?'] = df['Have you ever fallen victim to a phishing attack?'].map({'Yes': True, 'No': False})

# Clean 'Cybersecurity measures currently in place' column
measures_mapping = {
    'Antivirus Software': 'Antivirus Software',
    'Firewall': 'Firewall',
    'Two-Factor Authentication (2FA)': 'Two-Factor Authentication (2FA)',
    'Virtual Private Network (VPN)': 'Virtual Private Network (VPN)'
}
df['Cybersecurity measures currently in place'] = df['Cybersecurity measures currently in place'].apply(lambda x: ', '.join([measures_mapping.get(measure.strip(), '') for measure in x.split(',')]))

# Clean 'How often do you update your passwords?' column
df['How often do you update your passwords?'] = df['How often do you update your passwords?'].str.strip().str.capitalize()

# Clean 'Primary motivation for practicing good cybersecurity habits' column
df['Primary motivation for practicing good cybersecurity habits'] = df['Primary motivation for practicing good cybersecurity habits'].str.strip()

# Convert 'Receive regular cybersecurity training or awareness programs?' column to boolean
df['Receive regular cybersecurity training or awareness programs?'] = df['Receive regular cybersecurity training or awareness programs?'].map({'Yes': True, 'No': False})

# Clean 'Confidence in ability to identify and avoid online threats' column
df['Confidence in ability to identify and avoid online threats'] = df['Confidence in ability to identify and avoid online threats'].str.strip()

# Clean 'Age group' column
df['Age group'] = df['Age group'].str.strip()

# Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_cyberForm.csv', index=False)

# Display the cleaned DataFrame
print(df)
```
